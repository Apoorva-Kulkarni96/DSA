```python
def countfreq(arr):
    n = len(arr)
    visited = [False] * n
    result = []

    for i in range(n):
        if visited[i]:
            continue

        count = 1
        for j in range(i+1, n):
            if arr[i] == arr[j]:
                visited[j] = True
                count += 1

        result.append((arr[i], count))

    return result
```

```python
countfreq([10,15,10,5,10])
```

```python
def countfreq(arr):
    freq = {}
    n = len(arr)
    maxEle = 0
    minEle = 0
    maxFreq = 0
    minFreq = n

    for value in arr:
        freq[value] = freq.get(value, 0) + 1
    
    for key, value in freq.items():
        if value > maxFreq:
            maxEle = key
            maxFreq = value
        
        if value < minFreq:
            minEle = key
            minFreq = value

    return maxEle, minEle
```
