```python
class Solution:
    def printName(self, name , count, N):
        if count == N:
            return
        print(name)
        self.printName(name, count+1, N)

if __name__ == "__main__":
    sol = Solution()
    N = 5
    name = "Apoorva"

    sol.printName(name, 0, N)
```

Complexity Analysis
Time Complexity: O(N), we print our name exactly N times.
Space Complexity: O(N), stack space used for recursive calls.

```python
#Sum of N number (Parameterised)

class Solution:
    def SumOfNnumber(self,i,sum):

        if i < 1:
            print(sum)
            return
    
        self.SumOfNnumber(i-1,sum+i)
if __name__ == "__main__":
    sol = Solution()
    sum=0
    i = 3
    sol.SumOfNnumber(i,sum)
```

```python
# Forward Recursion

class Solution:
    def SumOfNnumber(self,i,sum,n):

        if i > n:
            print(sum)
            return
    
        self.SumOfNnumber(i+1,sum+i,n)
if __name__ == "__main__":
    sol = Solution()
    sum=0
    i = 0
    n =3
    sol.SumOfNnumber(i,sum,n)
```

```python
#Funtional Recursion

class Solution1:
    def SumOfNumber(self, n: int):
        if (n == 0):
            return 0
        
        return (n + self.SumOfNumber(n-1))
if __name__ == "__main__":
    sol = Solution1()
    n = 3
    sol.SumOfNumber(n)
```

```python
#Factorial of Number : Recursion

class Solution:
    def factorial(self, n):
        if n == 0:
            return 1
        return n * self.factorial(n-1)
if __name__ == "__main__":
    sol = Solution()

    n =5
    print(sol.factorial(n))
```

```python
class Solution:
    def ReverseArray(self, arr):
        p1 = 0
        p2 = len(arr)-1

        while p1 < p2:
            arr[p1], arr[p2]= arr[p2], arr[p1]

            p1= p1+1
            p2 =p2-1

        return arr
if __name__ == "__main__":
    sol = Solution()
    arr = [1,2,3,4,5]

    print(sol.ReverseArray(arr))
```

```python
def Palindrome(s):
    n = len(s)

    left, right = 0, n-1

    while left < right:
        if not s[left].isalnum():
            left += 1
        elif not s[right].isalnum():
            right -= 1
        elif s[left].lower() != s[right].lower():
            return False
        else:
            left += 1
            right -= 1
    return True
if __name__ == "__main__":
    str = "ABCCBA" 
    ans = Palindrome(str)

  
    if ans:
        print("Palindrome")
    else:
        print("Not Palindrome")
```

```python
def Palindrome(i,s):
    n = len(s)

    if i >= n//2:
        return True
    if s[i] != s[n-i-1]:
        return False
    return Palindrome(i+1, s)
if __name__ == "__main__":
    str = "Aafg" 
    print(Palindrome(0,str))
```

```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        cleaned_str= "".join([ch.lower() for ch in s if ch.isalnum()])
        n = len(cleaned_str)
        left, right = 0, n-1
        if s == "":
            return True
        while left < right:
            if cleaned_str[left] != cleaned_str[right]:
                return False
            else:
                left += 1
                right -= 1
        return True 
if __name__ == "__main__":
    s= "A man, a plaa canal: Panama"
    sol = Solution()
    print(sol.isPalindrome(s))
```
