### Recursion

```python
class Solution:
    def printName(self, name , count, N):
        if count == N:
            return
        print(name)
        self.printName(name, count+1, N)

if __name__ == "__main__":
    sol = Solution()
    N = 5
    name = "Apoorva"

    sol.printName(name, 0, N)
```

Complexity Analysis
Time Complexity: O(N), we print our name exactly N times.
Space Complexity: O(N), stack space used for recursive calls.

```python
#Sum of N number (Parameterised)

class Solution:
    def SumOfNnumber(self,i,sum):

        if i < 1:
            print(sum)
            return
    
        self.SumOfNnumber(i-1,sum+i)
if __name__ == "__main__":
    sol = Solution()
    sum=0
    i = 3
    sol.SumOfNnumber(i,sum)
```

```python
# Forward Recursion

class Solution:
    def SumOfNnumber(self,i,sum,n):

        if i > n:
            print(sum)
            return
    
        self.SumOfNnumber(i+1,sum+i,n)
if __name__ == "__main__":
    sol = Solution()
    sum=0
    i = 0
    n =3
    sol.SumOfNnumber(i,sum,n)
```

```python
#Funtional Recursion

class Solution1:
    def SumOfNumber(self, n: int):
        if (n == 0):
            return 0
        
        return (n + self.SumOfNumber(n-1))
if __name__ == "__main__":
    sol = Solution1()
    n = 3
    sol.SumOfNumber(n)
```

```python
#Factorial of Number : Recursion

class Solution:
    def factorial(self, n):
        if n == 0:
            return 1
        return n * self.factorial(n-1)
if __name__ == "__main__":
    sol = Solution()

    n =5
    print(sol.factorial(n))
```
