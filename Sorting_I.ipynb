### Sorting I

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n-1):
        min_index = i
        for j in range(i+1, n):
            if arr[j] < arr[min_index]:
                min_index=j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    print(*arr)
   
```

```python
def bubble_sort(arr):
    n = len(arr)
    
    for i in range(n-1):
        did_swap = False
        for j in range(0,n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1],arr[j]
                did_swap = True
        if not did_swap:
            break
    print(*arr)     
```

```python
def insertion(arr):

    n = len(arr)

    for i in range(1,n):
        key = arr[i]
        j = i-1

        while j >=0 and key < arr[j]:
            arr[j+1]  = arr[j]
            j -= 1
        arr[j+1] = key
    return arr
```

```python
insertion([5,3,2,8])
```
