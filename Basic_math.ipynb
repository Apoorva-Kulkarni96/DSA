```python
#Count all Digits of a Number

#Brute Force

class Solution:
    def countDigit(self, n):
      if n == 0:
        return 1
      n=abs(n)
      count=0
      while n>0:
        n//=10
        count=count+1
      return count
sol=Solution()
print(sol.countDigit(12344))

'''
Complexity Analysis

Time Complexity: O(log10N + 1) where N is the input number. The time complexity is determined by the number of digits in the input integer N. In the worst case when N is a multiple of 10 the number of digits in N is log10N + 1.

In the while loop we divide N by 10 until it becomes 0 which takes log10N iterations.
In each iteration of the while loop we perform constant time operations like division and increment the counter.
Space Complexity : O(1) as only a constant amount of additional memory for the counter regardless of size of the input number.

'''

#Optimal Solution

import math
class Solution:
    def countDigit(self, n):
      count= math.log10(n)+1
      return int(count)
sol=Solution()
print(sol.countDigit(12344))
```

```python
#Armstrong Numbers

class Solution:
    def isArmstrong(self, n):
      temp=n
      arm=0
      num=0
      d=len(str(n))
      while n > 0:
        num = n % 10
        arm=arm+(num**d)
        n//=10
      if temp == arm :
        return "Armstrong Number"
      else:
        return "Non Armstrong Number"

```

```python
#Check for Prime Number

#Brute Force

class Solution:

    def isPrime(self, n):
        count=0
        for i in range(1,n+1):
          if n%i==0:
            count=count+1
        if count == 2:
          return "Prime Number"
        else:
          return "Non Prime Number"


#Optimal Soution


class Solution:

    def isPrime(self, n):
        count=0
        N=int(math.sqrt(n))
        for i in range(1,N+1):
          if n%i==0:
            count=count+1

            if n%i != i:
              count=count+ 1
        if count == 2:
          return "Prime Number"
        else:
          return "Non Prime Number"

Sol=Solution()
print(Sol.isPrime(7))
```

```python
#GCD of Two Numbers

# Brute Force

class Solution:
    def GCD(self, n1, n2):

      for i in range(min(n1, n2),0,-1):
        if n1%i==0 and n2%i==0:
          return i


Sol=Solution()
print(Sol.GCD(12,20))


#Optimal Force ()


class Solution1:
    def GCD(self, n1, n2):
      while n1 > 0 and n2 >0:
        if n1 > n2:
          n1=n1%n2
        else:
          n2= n2%n1

      if n1==0:
        return n2
      return n1
Sol=Solution1()
print(Sol.GCD(12,20))
```
