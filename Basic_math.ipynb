```python
#Count all Digits of a Number

#Brute Force

class Solution:
    def countDigit(self, n):
      if n == 0:
        return 1
      n=abs(n)
      count=0
      while n>0:
        n//=10
        count=count+1
      return count
sol=Solution()
print(sol.countDigit(12344))

'''
Complexity Analysis

Time Complexity: O(log10N + 1) where N is the input number. The time complexity is determined by the number of digits in the input integer N. In the worst case when N is a multiple of 10 the number of digits in N is log10N + 1.

In the while loop we divide N by 10 until it becomes 0 which takes log10N iterations.
In each iteration of the while loop we perform constant time operations like division and increment the counter.
Space Complexity : O(1) as only a constant amount of additional memory for the counter regardless of size of the input number.

'''

#Optimal Solution

import math
class Solution:
    def countDigit(self, n):
      count= math.log10(n)+1
      return int(count)
sol=Solution()
print(sol.countDigit(12344))

'''
Code 1 (Iterative Division):
- Time Complexity: O(log10(n))   # loop runs once per digit
- Space Complexity: O(1)         # only a few variables used

Code 2 (Logarithmic Formula):
- Time Complexity: O(1)          # math.log10 runs in constant time
- Space Complexity: O(1)         # only a few variables used
'''
```

```python

#Reverse Digits of A Number

#Optimal

class Solution:
    def reverse(self, x: int) -> int:
        sign = -1 if x < 0 else 1
        rev=0
        rem=0
        x=abs(x)
        while x>0:
            rem = x%10
            rev=rev*10+rem
            x//=10
        rev =rev * sign

        if (rev > (2**31)-1) or (rev < (-2**31)):
            return 0
        return rev

'''
Reverse Function:
- Time Complexity: O(log10(x))   # loop runs once per digit
- Space Complexity: O(1)         # only a few variables used
'''
```

```python
#Armstrong Numbers

class Solution:
    def isArmstrong(self, n):
      temp=n
      arm=0
      num=0
      d=len(str(n))
      while n > 0:
        num = n % 10
        arm=arm+(num**d)
        n//=10
      if temp == arm :
        return "Armstrong Number"
      else:
        return "Non Armstrong Number"

'''
Time Complexity: O(log n)
Space Complexity: O(1)

'''
```

```python
#Check for Prime Number

#Brute Force

class Solution:

    def isPrime(self, n):
        count=0
        for i in range(1,n+1):
          if n%i==0:
            count=count+1
        if count == 2:
          return "Prime Number"
        else:
          return "Non Prime Number"


#Optimal Soution


class Solution:

    def isPrime(self, n):
        count=0
        N=int(math.sqrt(n))
        for i in range(1,N+1):
          if n%i==0:
            count=count+1

            if n//i != i:
              count=count+ 1
        if count == 2:
          return "Prime Number"
        else:
          return "Non Prime Number"

Sol=Solution()
print(Sol.isPrime(7))

'''
Prime Check (Brute Force):

Time Complexity:
- The loop runs from 1 to n.
- For each value, we perform a modulo operation (constant time).
- Therefore, total operations are proportional to n.

=> Time Complexity: O(n)

Space Complexity:
- Only a few variables are used (count, loop variable).
- No extra data structures are created.

=> Space Complexity: O(1)


Prime Check (Optimal √n Method):

Time Complexity:
- The loop runs from 1 to √n.
- Each iteration performs constant-time operations.
- In the worst case (when n is prime), the loop runs fully till √n.

=> Time Complexity: O(√n)

Space Complexity:
- Only constant variables are used (count, N, loop variable).
- No extra data structures are created.

=> Space Complexity: O(1)
'''
```

```python
#GCD of Two Numbers

# Brute Force

class Solution:
    def GCD(self, n1, n2):

      for i in range(min(n1, n2),0,-1):
        if n1%i==0 and n2%i==0:
          return i


Sol=Solution()
print(Sol.GCD(11,13))


#Optimal Force ()


class Solution1:
    def GCD(self, n1, n2):
      while n1 > 0 and n2 >0:
        if n1 > n2:
          n1=n1%n2
        else:
          n2= n2%n1

      if n1==0:
        return n2
      return n1
Sol=Solution1()
print(Sol.GCD(11,13))


'''
GCD (Brute Force):

Time Complexity:
- The loop starts from min(n1, n2) and goes down to 1.
- In the worst case (when GCD is 1), the loop checks all values.
- Therefore, total iterations are proportional to min(n1, n2).

=> Time Complexity: O(min(n1, n2))

Space Complexity:
- No extra data structures are used.
- Only loop variable is used.

=> Space Complexity: O(1)

GCD (Optimal – Euclidean Algorithm):

Time Complexity:
- Uses the property:
      gcd(a, b) = gcd(b, a % b)
- In each iteration, the larger number reduces significantly.
- The values shrink quickly due to modulo operation.
- The number of steps is proportional to log(min(n1, n2)).

=> Time Complexity: O(log(min(n1, n2)))

Space Complexity:
- Only constant variables are used.
- No extra data structures are created.

=> Space Complexity: O(1)


'''
```
