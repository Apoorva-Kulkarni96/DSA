### 1. Find the Largest element in an array

#### Optimal Approach

```python
arr = [2,5,1,3,0]

def largest_element(arr):
    max_num = arr[0]

    for i in range(len(arr)):
        if arr[i] > max_num:
            max_num = arr[i]
    return max_num
```

### 2. Find Second Smallest and Second Largest Element in an array

#### Brute Approach

```python
def second_largest_second_largest(arr,n):
    if n == 0 or n == 1:
        print(-1,-1)
        return
    arr.sort()

    sec_largest = arr[n-2] # arr[-2]
    sec_smallest = arr[1]

    return sec_largest, sec_smallest
arr = [2,5,1,3,0]

second_largest_second_largest(arr, len(arr))
```

Time Complexity: O(n logn)

Space Complexity: O(1)

#### Better Approach

```python
def second_largest_and_second_largest(arr,n):
    if n == 0 or n == 1:
        return -1,-1

    small = float("inf")
    second_small = float("inf")
    largest = float("-inf")
    second_largest = float("-inf")

    for i in range(n):
        small = min(arr[i], small)
        largest = max(arr[i], largest)

    for i in range(n):
        if arr[i] < second_small and arr[i] != small:
            second_small = arr[i]
        if arr[i] > second_largest and arr[i] != largest:
            second_largest = arr[i]

    return second_small, second_largest


if __name__ == "__main__":

    arr=[2,5,1,3,0]
    n = len(arr)

    second_small, second_large = second_largest_and_second_largest(arr, n)
    print(second_small, second_large)
```

Time Complexity: O(n)

Space Complexity: O(1)

#### Optimal Approach

```python
def second_smallest(arr,n):
    if n < 2:
        return -1
    
    small = float("inf")
    second_small = float("inf")

    for i in range(n):
        if arr[i] < small:
            second_small = small
            small = arr[i]
        elif arr[i] < second_small and arr[i] != small:
            second_small = arr[i]

    return second_small

def second_largest(arr,n):
    if n < 2:
        return -1
    
    large = float("-inf")
    second_large = float("-inf")

    for i in range(n):
        if arr[i] > large:
            second_large = large
            large = arr[i]
        elif arr[i] > second_large and arr[i] != large:
            second_large = arr[i]

    return second_large


if __name__ == "__main__":
    arr = [1, 2, 4, 7, 7, 5]  # Array of elements
    n = len(arr)  # Size of the array

    # Find the second smallest and second largest elements
    sS = second_smallest(arr, n)
    sL = second_largest(arr, n)

    print(sS,sL)
```

Time Complexity: O(n)

Space Complexity: O(1)

### Check if an Array is Sorted

#### Brute Approach

```python
def isSorted(arr,n):
    for i in range(n):
        for j in range(i+1,n):
            if arr[i] > arr[j]:
                return False
    return True

if __name__ == "__main__":
    arr = [1,2,7,4,5,6]
    n = len(arr)
    ans = isSorted(arr,n)

    if ans:
        print("Sorted")
    else:
        print("Unsorted")
```

Time Complexity: O(n^2)

Space Complexity: O(1)

#### Optimal Approach

```python
def isSorted(arr,n):
    for i in range(1,n):
        if arr[i] < arr[i-1]:
            return False
    return True


if __name__ == "__main__":
    arr = [1,2,4,5,6]
    n = len(arr)
    ans = isSorted(arr,n)

    if ans:
        print("Sorted")
    else:
        print("Unsorted")
```

Time Complexity: O(n)

Space Complexity: O(1)

### Check if an Array is Sorted and Rotate

#### Brute Force

```python
class Solution:
    def check(self, nums):
        n = len(nums)
        for k in range(n):
            if self.isSorted(self.rotate(nums[:],k)) == True:
                return True
        return False
            

    
    def isSorted(self, nums):
        for i in range(len(nums)-1):
            if nums[i] > nums[i+1]:
                return False
        return True
    
    def rotate(self, nums,k):
        n = len(nums)
        k  = k%n
        nums[k:] = reversed(nums[k:])
        nums[:k] = reversed(nums[:k])
        nums.reverse()
        return nums
```

```python
sol = Solution()
print(sol.check([3,4,5,1,2]))  # True
print(sol.check([2,1,3,4]))    # False
print(sol.check([1,2,3]))      # True
```

Time Complexity: O(n^2)

Space Complexity: O(n)

### Remove Duplicates in-place from Sorted Array

#### Brute Approach

```python
class Solution:
    def remove_duplicate(self, nums):

        seen = set()
        index = 0

        for num in nums:
            if num not in seen:
                seen.add(num)
                nums[index] = num

                index += 1
        return index
    
if __name__ == "__main__":
    sol = Solution()
    nums = [0,0,1,1,1,2,2,3,3,4]
    k = sol.remove_duplicate(nums)

    print(nums[:k])
```

Time Complexity: O(n)

Space Complexity: O(n)

#### Optimal Approach 

```python
class Solution:
    def remove_duplicate(self, nums):
        i = 0
        n =len(nums)

        for j in range(1, n):
            if nums[i] != nums[j]:
                i += 1
                nums[i] = nums[j]
        return i+1
if __name__ == "__main__":
    sol = Solution()
    nums = [0,0,1,1,1,2,2,3,3,4]
    k = sol.remove_duplicate(nums)

    print(nums[:k])
```

Time Complexity: O(n)

Space Complexity: O(1)


### Left Rotate the Array by One



### Brute Approach

```python
def left_rotate(arr):
    new_arr=[]
    for i in range(1,n):
        new_arr.append(arr[i])
    new_arr.append(arr[0])
    return new_arr
```

Time Complexity: O(n)

Space Complexity: O(n)

### Optimal Approach

```python


def left_rotate(arr):
    n = len(arr)
    first = arr[0]
    for j in range(n-1):
        arr[j] = arr[j+1]
    arr[n-1] = first
    return arr
```

Time Complexity: O(n)

Space Complexity: O(1)

### Rotate array by K elements

### Brute Force

```python
def rotate_k_element(nums, direction, n, k):
    k = k%n
    if direction == "left":
        return nums[k:] + nums[:k]
    else:
        return nums[-k:] + nums[:-k]
        
```

```python
rotate_k_element([1, 2, 3, 4, 5,6,7], "left", 7, 2)
```

```python
rotate_k_element([1, 2, 3, 4, 5,6,7], "right", 7, 2)
```

Time Complexity: O(n)

Space Complexity: O(n)

### Optimal Approach

```python
def rotate_k_element_optimal(nums, direction,n, k):
    k = k % n
    if direction == "left":
        nums[k:] = reversed(nums[k:])
        nums[:k] = reversed(nums[:k])
        nums.reverse()
        return nums
    else:
        nums[-k:] = reversed(nums[-k:])
        nums[:-k] = reversed(nums[:-k])
        nums.reverse()
        return nums

    
```

Time Complexity: O(n)

Space Complexity: O(1)

### Move all Zeros to the end of the array

#### Brute Force

```python
def move_zero(arr):
    n = len(arr)
    temp = []
    for i in range(n):
        if arr[i] != 0:
            temp.append(arr[i])
    length_non_zero = len(temp)
    for j in range(len(temp)):
        arr[j] = temp[j]

    for k in range(length_non_zero,n):
        arr[k]=0
    return arr
```

```python
def move_zero_prefilled_zero(arr):
    temp = [0] * len(arr)
    index = 0

    for num in arr:
        if num != 0:
            temp[index] = num
            index += 1

    arr[:] = temp

    return arr
```

```python
move_zero_prefilled_zero([1 ,0 ,2 ,3 ,0 ,4 ,0 ,1])
```

Time Complexity: O(n)

Space Complexity: O(n)

#### Optimal Solution

```python
class Solution:
    def movezeros(self, nums):
        j = -1

        for i in range(len(nums)):
            if nums[i] == 0:
                j = i
                break

        if j == -1:
            return 
        
        for i in range(j+1, len(nums)):
            if nums[i] != 0:
                nums[i], nums[j] = nums[j], nums[i]
                j += 1
```

```python
sol = Solution()
nums = [0, 1, 0, 3, 12]
sol.movezeros(nums)
print(nums)
```

Time Complexity: O(n)

Space Complexity: O(n)

### Find the missing number

#### Brute Approach (Linear Search)

```python
def MissingNum(arr):
    n = len(arr) + 1

    for i in range(1,n+1):
        found = False
        for j in range(n-1):
            if arr[j] == i:
                found = True
                break
        
        if not found:
            return i
MissingNum([8, 2, 4, 5, 3, 7, 1])
```

Time Complexity: O(n^2)

Space Complexity: O(1)

### Using HashMap

```python
def MissingNum(arr):
    n = len(arr) + 1

    hash = [0] * (n+1)

    for i in range(n-1):
        hash[arr[i]] += 1

    for i in range(1, n+1):
        if hash[i] == 0:
            return i
    return -1

MissingNum([8, 2, 4, 5, 3, 7, 1])
```

Time Complexity: O(n)

Space Complexity: O(n)

#### Use Sum of Natural Number

```python
def MissingNum(arr):
    n = len(arr) + 1
    totalsum = 0
    expected =0
    for i in range(n-1):
        totalsum += arr[i]
    
    for j in range(1,n+1):
        expected += j
    
    return expected - totalsum

MissingNum([8, 2, 4, 5, 3, 7, 1])
```

Time Complexity: O(n)

Space Complexity: O(1)

#### Using XOR

```python
def MissingNum(arr):
    xor1 = 0
    xor2 = 0
    n = len(arr) + 1
    for i in range(n-1):
        xor2 ^= arr[i]

    for j in range(1, n+1):
        xor1 ^= j
    
    return xor1^xor2

MissingNum([8, 2, 4, 5, 3, 7, 1]) 
```

Time Complexity: O(n)

Space Complexity: O(1)

### Count Maximum Consecutive One's in the array

```python
class Solution():
    def MaxConsecutiveOnes(self,arr):
        count = 0
        maximum = 0
        for i in range(len(arr)):
            if arr[i] == 1:
                count += 1
            else:
                count = 0
            maximum = max(maximum, count)
        return maximum

if __name__ == "__main__":
    sol = Solution()
    arr = [1, 1, 0, 1, 1, 1]
    a= sol.MaxConsecutiveOnes(arr)
    print(a)
```

Time Complexity: O(n)

Space Complexity: O(1)

### Find the number that appears once, and the other numbers twice


### Brute Force

```python
def getSingleElement(arr):
    n = len(arr)

    for i in range(n):
        num = arr[i]
        count = 0

        for j in range(n):
            if arr[j] == num:
                count += 1
        if count == 1:
            return num
    return -1
getSingleElement([4, 1, 2, 1, 2])
```

Time Complexity: O(n^2)

Space Complexity: O(1)

### Dictionary

```python
def getsingleelement(arr):

    freq = {}

    for i in range(len(arr)):
        freq[arr[i]] = freq.get(arr[i], 0) + 1

    for key, value in freq.items():
        if value == 1:
            return key
    return -1
```

```python
getsingleelement([4,1,1,2,2])
```

Time Complexity: O(n)

Space Complexity: O(n)

### Hash table

```python
def singleElement(arr):
    n = len(arr)
    maxi  = max(arr)

    hash_arr = [0] * (maxi+1)

    for num in arr:
        hash_arr[num] += 1
    
    for num in arr:
        if hash_arr[num] == 1:
            return num
    return -1
```

```python
singleElement([4,1,2,1,2])
```

Time Complexity: O(n)

Space Complexity: O(maxi+1)
